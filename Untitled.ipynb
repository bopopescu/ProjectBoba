```python
###### SETUP #########
keys = {}
with open("key.json","r") as f:
    keys = json.loads(f.read())
API_key = keys["yelp_API"]
search_URL = "https://api.yelp.com/v3/businesses/search"
business_details_URL = "https://api.yelp.com/v3/businesses/"
authorization = "Bearer " + API_key
authorizer = {"Authorization":authorization}
```

```python
payload = {"location": "NYC","categories": "bubbletea", "limit":50}
results = re.get(search_URL, params = payload, headers = authorizer)
```
